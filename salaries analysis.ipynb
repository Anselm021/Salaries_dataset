```python
# Ensure 'salary_in_usd' is numeric
salaries['salary_in_usd'] = pd.to_numeric(salaries['salary_in_usd'], errors='coerce')

# Drop rows where salary could not be converted to numeric values
salaries = salaries.dropna(subset=['salary_in_usd'])
```

```python

# Group by job_title and calculate the average salary
average_salary_by_job = salaries.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Sort by salary in descending order
top_15_jobs = average_salary_by_job.sort_values(by='salary_in_usd', ascending=False).head(15)
top_15_jobs
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='salary_in_usd', y='job_title', data=top_15_jobs, palette='viridis')

# Add titles and labels
plt.title('Top 15 Job Titles by Average Salary in USD', fontsize=14)
plt.xlabel('Average Salary (USD)', fontsize=12)
plt.ylabel('Job Title', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting Experience Level by Company Size
plt.figure(figsize=(10, 6))
sns.countplot(data=salaries, x='company_size', hue='experience_level', palette='Set1')

# Add titles and labels
plt.title('Experience Level by Company Size', fontsize=14)
plt.xlabel('Company Size', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting Experience Level by Work Year
plt.figure(figsize=(10, 6))
sns.countplot(data=salaries, x='work_year', hue='experience_level', palette='Set2')

# Add titles and labels
plt.title('Experience Level by Work Year', fontsize=14)
plt.xlabel('Work Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by company_location and calculate average salary in USD
top_locations_salary = salaries.groupby('company_location')['salary_in_usd'].mean().reset_index()

# Sort by average salary in descending order and select the top 10 locations
top_10_locations = top_locations_salary.sort_values(by='salary_in_usd', ascending=False).head(10)

# Plotting the top 10 locations by average salary
plt.figure(figsize=(12, 6))
sns.barplot(x='salary_in_usd', y='company_location', data=top_10_locations, palette='coolwarm')

# Add titles and labels
plt.title('Top 10 Locations by Highest Average Salary in USD', fontsize=14)
plt.xlabel('Average Salary (USD)', fontsize=12)
plt.ylabel('Company Location', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 2. Experience level and employment type by work year
plt.figure(figsize=(10, 8))
sns.countplot(data=salaries, x='work_year', hue='experience_level', palette='coolwarm', dodge=True)
plt.title('Experience Level by Work Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Experience Level')
plt.show()

plt.figure(figsize=(10, 8))
sns.countplot(data=salaries, x='work_year', hue='employment_type', palette='Set2', dodge=True)
plt.title('Employment Type by Work Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Employment Type')
plt.show()
```

```python
# 3. Company size by year
company_size_by_year = salaries.groupby(['work_year', 'company_size']).size().reset_index(name='count')
plt.figure(figsize=(10, 8))
sns.barplot(data=company_size_by_year, x='work_year', y='count', hue='company_size', palette='Set3')
plt.title('Company Size by Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Company Size')
plt.show()
```

```python
# 4. Top 5 job titles by year
def get_top_5_jobs_by_year(salaries):
    top_jobs_by_year = []
    for year in salaries['work_year'].unique():
        yearly_data = salaries[salaries['work_year'] == year]
        top_5_jobs = (yearly_data['job_title']
                      .value_counts()
                      .head(5)
                      .reset_index())
        top_5_jobs['work_year'] = year
        top_5_jobs.columns = ['job_title', 'count', 'work_year']
        top_jobs_by_year.append(top_5_jobs)
    return pd.concat(top_jobs_by_year, ignore_index=True)

top_jobs_by_year = get_top_5_jobs_by_year(salaries)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_jobs_by_year, x='work_year', y='count', hue='job_title', palette='Paired')
plt.title('Top 5 Job Titles by Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Job Title')
plt.show()
```

```python
# 5. Top 3 company locations by year
def get_top_3_locations_by_year(salaries):
    top_locations_by_year = []
    for year in salaries['work_year'].unique():
        yearly_data = salaries[salaries['work_year'] == year]
        top_3_locations = (yearly_data['company_location']
                           .value_counts()
                           .head(3)
                           .reset_index())
        top_3_locations['work_year'] = year
        top_3_locations.columns = ['company_location', 'count', 'work_year']
        top_locations_by_year.append(top_3_locations)
    return pd.concat(top_locations_by_year, ignore_index=True)

top_locations_by_year = get_top_3_locations_by_year(salaries)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_locations_by_year, x='work_year', y='count', hue='company_location', palette='Spectral')
plt.title('Top 3 Company Locations by Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Company Location')
plt.show()
```

```python
# 6. Predict salary trends for the next 5 years
# Prepare the data for the model
from sklearn.linear_model import LinearRegression
salary_trends = salaries.groupby('work_year')['salary_in_usd'].mean().reset_index()
X = salary_trends['work_year'].values.reshape(-1, 1)
y = salary_trends['salary_in_usd'].values

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the next 5 years
future_years = np.array(range(salaries['work_year'].max() + 1, salaries['work_year'].max() + 6)).reshape(-1, 1)
predicted_salaries = model.predict(future_years)

# Combine past and predicted data
future_data = pd.DataFrame({'work_year': future_years.flatten(), 'salary_in_usd': predicted_salaries})
combined_data = pd.concat([salary_trends, future_data], ignore_index=True)

# Plot the trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_data, x='work_year', y='salary_in_usd', marker='o', label='Salary Trend')
plt.title('Salary Trends (Past and Predicted)')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.legend()
plt.show()
```

```python
# 6. Predict salary trends for the next 5 years based on top 5 job titles with highest average salary
# Identify the top 5 job titles with the highest average salary
top_5_jobs = (salaries.groupby('job_title')['salary_in_usd']
               .mean()
               .sort_values(ascending=False)
               .head(5)
               .index)

# Filter the dataset for the top 5 job titles
filtered_salaries = salaries[salaries['job_title'].isin(top_5_jobs)]

# Group by year and calculate the average salary for these job titles
salary_trends_top_jobs = (filtered_salaries.groupby('work_year')['salary_in_usd']
                          .mean()
                          .reset_index())

X_top_jobs = salary_trends_top_jobs['work_year'].values.reshape(-1, 1)
y_top_jobs = salary_trends_top_jobs['salary_in_usd'].values

# Train a linear regression model
model_top_jobs = LinearRegression()
model_top_jobs.fit(X_top_jobs, y_top_jobs)

# Predict the next 5 years
future_years_top_jobs = np.array(range(salaries['work_year'].max() + 1, salaries['work_year'].max() + 6)).reshape(-1, 1)
predicted_salaries_top_jobs = model_top_jobs.predict(future_years_top_jobs)

# Combine past and predicted data
future_data_top_jobs = pd.DataFrame({'work_year': future_years_top_jobs.flatten(), 'salary_in_usd': predicted_salaries_top_jobs})
combined_data_top_jobs = pd.concat([salary_trends_top_jobs, future_data_top_jobs], ignore_index=True)

# Plot the trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_data_top_jobs, x='work_year', y='salary_in_usd', marker='o', label='Salary Trend for Top 5 Job Titles')
plt.title('Salary Trends for Top 5 Job Titles (Past and Predicted)')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.legend()
plt.show()
```
