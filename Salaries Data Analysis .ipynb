**TOP 5 JOB TITLES BY YEAR**

```python
# Top 5 job titles by year
def get_top_5_jobs_by_year(salaries):
    top_jobs_by_year = []
    for year in salaries['work_year'].unique():
        yearly_data = salaries[salaries['work_year'] == year]
        top_5_jobs = (yearly_data['job_title']
                      .value_counts()
                      .head(5)
                      .reset_index())
        top_5_jobs['work_year'] = year
        top_5_jobs.columns = ['job_title', 'count', 'work_year']
        top_jobs_by_year.append(top_5_jobs)
    return pd.concat(top_jobs_by_year, ignore_index=True)

top_jobs_by_year = get_top_5_jobs_by_year(salaries)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_jobs_by_year, x='work_year', y='count', hue='job_title', palette='Paired')
plt.title('Top 5 Job Titles by Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Job Title')
plt.show()
```

**TOP 3 COMPANY LOCATIONS BY YEAR**

```python
# Top 3 company locations by year
def get_top_3_locations_by_year(salaries):
    top_locations_by_year = []
    for year in salaries['work_year'].unique():
        yearly_data = salaries[salaries['work_year'] == year]
        top_3_locations = (yearly_data['company_location']
                           .value_counts()
                           .head(3)
                           .reset_index())
        top_3_locations['work_year'] = year
        top_3_locations.columns = ['company_location', 'count', 'work_year']
        top_locations_by_year.append(top_3_locations)
    return pd.concat(top_locations_by_year, ignore_index=True)

top_locations_by_year = get_top_3_locations_by_year(salaries)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_locations_by_year, x='work_year', y='count', hue='company_location', palette='Spectral')
plt.title('Top 3 Company Locations by Year')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.legend(title='Company Location')
plt.show()
```

**Predict salary trends for the next 5 years**

```python
# Predict salary trends for the next 5 years
# Prepare the data for the model
from sklearn.linear_model import LinearRegression
salary_trends = salaries.groupby('work_year')['salary_in_usd'].mean().reset_index()
X = salary_trends['work_year'].values.reshape(-1, 1)
y = salary_trends['salary_in_usd'].values

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the next 5 years
future_years = np.array(range(salaries['work_year'].max() + 1, salaries['work_year'].max() + 6)).reshape(-1, 1)
predicted_salaries = model.predict(future_years)

# Combine past and predicted data
future_data = pd.DataFrame({'work_year': future_years.flatten(), 'salary_in_usd': predicted_salaries})
combined_data = pd.concat([salary_trends, future_data], ignore_index=True)

# Plot the trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_data, x='work_year', y='salary_in_usd', marker='o', label='Salary Trend')
plt.title('Salary Trends (Past and Predicted)')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.legend()
plt.show()
```
